```python
import sys
import os

# Get the user's home directory
home_dir = os.path.expanduser("~")

# Construct the path to the user's package directory
user_package_dir = os.path.join(home_dir, ".local", "lib", "python3.10", "site-packages")  # Adjust "python3.10" to your Python version

# Add the user package directory to PYTHONPATH
sys.path.append(user_package_dir)

import numpy as np
import pandas as pd
import gensim
import os
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
import nltk
nltk.download('punkt_tab')
```

```python
story = []
for filename in os.listdir('data'):
    if filename == '.ipynb_checkpoints':
      pass
    f = open(os.path.join('data',filename))
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
len(story)
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2)
```

```python
model.build_vocab(story)
```

```python
model.corpus_count
```

```python
model.epochs
```

```python

model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```
